```python
import numpy as np

# data
X = np.array([1, 2, 4, 6, 8])  # feature
y = np.array([2, 5, 6, 9, 11])  # target

# Parameters
a = 0
b = 0
step_size = 0.01
threshold = 0.01
iterations = 0
```

```python
# Gradient Descent
while True:
    #linear regression model
    y_pred = a + b * X
    #Calculate the error
    error = y_pred - y
    da = np.sum(error)
    db = np.sum(error * X)
    #update the parameter
    a -= step_size * da / len(X)
    b -= step_size * db / len(X)
    
    gradient_magnitude = np.sqrt(da**2 + db**2)
    iterations += 1  # Increment the iteration count
    
    # Break the loop if the gradient magnitude is less than the threshold
    if gradient_magnitude < threshold:
        break

# Print the results
print(f"Converged after {iterations} iterations with a = {a}, b = {b}")
```

```python
# Plotting
import matplotlib.pyplot as plt
plt.scatter(X, y, color='red', label='Actual data')
plt.plot(X, a + b * X, color='blue', label='Fitted line')
plt.title('Linear Regression Fit')
plt.xlabel('X (feature)')
plt.ylabel('y (target)')
plt.legend()
plt.show()
```
